```python
# As the statistical measure shows the data irrelevancy, i will extract the relevant data which will drive valuable insights. Lets Form a new Data set
```

```python
data.rename(columns = {'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'State','region_txt':'Region','attacktype1_txt':'Attack_type','target1':'Target_Type','nkill':'Killed','nwound':'Wounded','summary':'Summary','gname':'Terror_Group','weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

```python
New_data = data[['Year','Month','Day','Country','State','Region','city','latitude','longitude','Attack_type','Killed','Wounded','Summary','Terror_Group','Target_Type','Weapon_type','Motive']]
```

```python
# After transforming new data set lets get deep dive into it to uncover missing values and fixing it as well.
```

```python
New_data.describe()
```

```python
New_data.isnull().sum()
```

```python
New_data['State'].fillna(0,inplace=True)
New_data['city'].fillna(0,inplace=True)
New_data['latitude'].fillna(0,inplace=True)
New_data['longitude'].fillna(0,inplace=True)
New_data['Killed'].fillna(0,inplace=True)
New_data['Wounded'].fillna(0,inplace=True)
New_data['Target_Type'].fillna(0,inplace=True)
New_data['Summary'].fillna(0,inplace=True)
New_data['Motive'].fillna(0,inplace=True) 
New_data['Terror_Group'].fillna(0,inplace=True)
New_data['Weapon_type'].fillna(0,inplace=True)
```

```python
Year=New_data.Year.value_counts().to_dict()
rate=((Year[2017]-Year[1970])/Year[2017])*100
print('A total of', Year[1970],'attacks occured in 1970 and a total of',Year[2017],'attacks occured in the year 2017.')
print('The total number of attacks from 1970 to 2017 has increased by',np.round(rate,0),'percent.')
```
